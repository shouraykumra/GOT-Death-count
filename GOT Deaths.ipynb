# Game of Thrones Deaths

#### Every on screen death in the Game of Thrones TV show is in this data set - details will include who, where, what and when with regards to each kill.

COLUMN NAME

name: The victim's name

allegiance: The victim's house they were associated with

season: The season of the tv show that the death occurred

episode: The episode of the tv show that the death occurred

location: The location the death occured

killer: The killer's name

killers_house: The killer's house they were associated with

method: How the killer killed their victim

```python
df=pd.read_excel('game-of-thones-deaths.xlsx')
```

```python
df
```

As 1st row here contains column names so we can convert this first row into column names with the following code.

```python
new_header=df.iloc[0] #selecting first row.
df=df[1:] # treats dataset from second row onwards.
df.columns=new_header
```

```python
df
```

```python
df['Killer'].value_counts()
```

Forming a list to store counts of each Killer's kill streak.

```python
list_kills=[]
list_kills.append(df['Killer'].value_counts())
list_kills[0][1]
```

For plotting a plotly.groupby we need to have number of kills the killers made as it needs to have some value to plot on.

```python
list_kills_2=[]
for i in range(129):
    list_kills_2.append(list_kills[0][i])
print(list_kills_2)
```

Assessing the index values of killers value counts.

https://stackoverflow.com/questions/36973387/accessing-first-column-of-pandas-value-counts

```python
index_val=df['Killer'].value_counts().index
```

```python
index_val_list=[]
for i in index_val:
    index_val_list.append(i)
print(index_val_list)
```

## Top Killers

Daenerys being the top killer has the highest count of 1008 people. No wonder how she could have ruled the world with no people in it! ;(

```python
import plotly.express as px
fig=px.histogram(df, x='Killer')
fig.show()
```

##### using both the list to plot the figure.

```python
import plotly.io as pio
data = [dict(
  type = 'scatter',
  x = index_val_list,
  y = list_kills_2,
  mode = 'markers',
  transforms = [dict(
    type = 'groupby',
    groups = index_val_list,
  )]
)]

fig_dict = dict(data=data)
pio.show(fig_dict, validate=False)
```
